```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
direct_link = 'https://assets.datacamp.com/production/repositories/655/datasets/2a1f3ab7bcc76eef1b8e1eb29afbd54c4ebf86f2/eurovision-2016.csv'
```

```python
df = pd.read_csv(direct_link)
```

```python
df
```

```python
country_names = df.iloc[:,0]
```

```python
sample = df.iloc[:,2:]
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

# Calculate the linkage: mergings
mergings = linkage(sample, method = 'single')

dendrogram(
    mergings, 
    labels = country_names,
    leaf_rotation=90,
    leaf_font_size=6)
```
